```python
import pandas as pd
from sklearn.metrics import r2_score, mean_absolute_error
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("appwf_res.csv")
```

```python
colors = {"neut": "black", "sdn": "blue", "ssv": "red"}

for swp in df["sweep"].unique():
    subdf = df[df["sweep"] == swp]

    plt.scatter(x=subdf["estimated_s_mean"], y=subdf["s_val"])
    plt.xlim(-0.1, .25)
    plt.ylim(0, .25)
    plt.annotate("r2 = {:.3f}".format(r2_score(subdf["s_val"], subdf["estimated_s_mean"])), (0.18, 0.04))
    plt.annotate("MAE = {:.3f}".format(mean_absolute_error(subdf["s_val"], subdf[f"estimated_s_mean"])), (0.18, 0.02))

    plt.title(f"WFApprox {swp.upper()} True vs Estimated s") 
    plt.xlabel("Estimated s") 
    plt.ylabel("True s") 
    plt.savefig(f"{swp}_true_v_est_wfapp.pdf")
    plt.show()
    plt.clf()
```
